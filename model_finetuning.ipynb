```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Initialize a text file to store all PDFs' content for LLM fine-tuning
output_file_path = "fine_tuning_data.txt"
with open(output_file_path, "w") as f:
    f.write("")  # Clear file if it exists

# Loop through each PDF file in the list
for pdf_path in pdf_list:
    # Open PDF document
    pdf_document = fitz.open(pdf_path)
    file_name = pdf_path.split('/')[-1]  # Get the last part of the path
    pdf_name = file_name.rsplit('.', 1)[0]  # Remove the file extension

    # Initialize a list to collect formatted text for this PDF
    formatted_text = []

    # Loop through each page in the PDF
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)

        # Extract and add text from the page
        text = page.get_text()
        if text:
            formatted_text.append(text.strip())

        # Extract text from images if any are present
        for img_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image = Image.open(io.BytesIO(image_bytes))

            # Perform OCR on the image and add extracted text
            image_text = pytesseract.image_to_string(image)
            if image_text:
                formatted_text.append(image_text.strip())

    # Combine all extracted text for this PDF into a single string, with spacing for readability
    pdf_text = "\n\n".join(formatted_text)

    # Append the extracted text for this PDF to the output file
    with open(output_file_path, "a") as f:
        f.write(f"\n\n--- Content from {pdf_name} ---\n\n")  # Add a header for each PDF
        f.write(pdf_text)

    print(f"Text data extracted and appended for {pdf_name}")

print(f"All PDFs processed and data saved to {output_file_path} for LLM fine-tuning.")
```

```python
import re

# Preprocessing function
def preprocess_text(text):
    # Remove unwanted symbols like bullets or special characters
    text = re.sub(r"[♦√•#■❏■]", "", text)

    # Remove multiple spaces and replace with single space
    text = re.sub(r"\s+", " ", text)

    # Normalize multiple newlines to a maximum of two newlines
    text = re.sub(r"\n{2,}", "\n\n", text)

    # Optional: Lowercase conversion (useful for some models)
    # text = text.lower()

    # Strip extra whitespace from the beginning and end
    return text.strip()

# Read the extracted text data
with open("fine_tuning_data.txt", "r") as f:
    raw_text = f.read()

# Preprocess the text
cleaned_text = preprocess_text(raw_text)

# Save the cleaned text to a new file for fine-tuning
with open("cleaned_fine_tuning_data.txt", "w") as f:
    f.write(cleaned_text)

print("Text data preprocessed and saved to cleaned_fine_tuning_data.txt")
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer, Trainer, TrainingArguments, TextDataset, DataCollatorForLanguageModeling

# Step 1: Load the Pre-trained Model and Tokenizer
model_name = "gpt2"  # You can replace this with another model if desired
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Set pad_token_id to eos_token_id to avoid warnings
tokenizer.pad_token = tokenizer.eos_token

# Step 2: Prepare the Dataset for Fine-Tuning
def load_dataset(file_path, tokenizer, block_size=128):
    # Create a TextDataset compatible with Hugging Face transformers
    return TextDataset(
        tokenizer=tokenizer,
        file_path=file_path,
        block_size=block_size  # Block size defines token length per sample; adjust based on memory
    )

# Load the cleaned text dataset
dataset = load_dataset("cleaned_fine_tuning_data.txt", tokenizer)
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)  # mlm=False since GPT-2 is not masked

# Step 3: Define Training Arguments
training_args = TrainingArguments(
    output_dir="./fine_tuned_model",
    overwrite_output_dir=True,
    num_train_epochs=3,  # Adjust as needed
    per_device_train_batch_size=2,  # Adjust based on GPU memory
    save_steps=500,
    save_total_limit=2,
    logging_dir='./logs',
    logging_steps=10
)

# Step 4: Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=dataset
)

# Step 5: Fine-Tune the Model
trainer.train()

# Step 6: Save the Fine-Tuned Model
model.save_pretrained("fine_tuned_emergency_bot")
tokenizer.save_pretrained("fine_tuned_emergency_bot")

print("Model fine-tuning complete and saved as 'fine_tuned_emergency_bot'")
```
